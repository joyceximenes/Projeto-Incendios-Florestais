```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('/content/drive/MyDrive/Study analise de dados/Data Viking/Projeto Incendios Florestais/Dados_Indendio.csv', encoding='latin-1')
dados.head()
```

```python
# contagem de nulos
dados.isnull().sum()
```

```python
plt.title('Total de nulos')
sns.heatmap(dados.isnull(), cbar=False);
```

```python
# campos unicos
dados.nunique()
```

```python
# analise por ano

analise = dados.groupby(by = ['year']).sum().reset_index()
analise.head()

plt.figure(figsize=(14,8))
plt.title('Relação do numero de incendios no Brasil por ano', fontsize = 15)
sns.lineplot(data = analise, x='year', y = 'number',  estimator='sum', lw=2, color='#ff5555', alpha=0.85 );

# dando nome pros eixos
plt.xlabel('Periodo')
plt.ylabel('Numero');
```

```python
# analise por mes

analise2 = dados.groupby(by = ['year','month']).sum().reset_index()
analise2.head()

plt.figure(figsize=(14,8))
plt.title('Incendios por mês', fontsize = 15)
sns.boxplot(data=analise2, x = 'month', y ='number', palette='coolwarm', saturation=1, width=0.9, linewidth=2,
           order=['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']);
# eixos
plt.xlabel('Mes')
plt.ylabel('Numero de incidencia');
```

```python
# analise por estado dos incendios
Analise_03 = dados.groupby( by=['state'] ).sum().reset_index()[['state', 'number']].sort_values( 'number', ascending=False ) # ordenando do maior para o menor
Analise_03.head()

# Tamanho
plt.figure( figsize=(15, 5) )

# Grafico
plt.title( 'Estados com maior número de incêndios', loc='left', fontsize=14 )

# Grafico
plt.bar( Analise_03.state, Analise_03['number'], color='#f44e3f' )

# Labels
plt.ylabel('Quantidade')
plt.xticks( rotation=80 );
```

```python
# listando estados com maior incidencia 
Lista_TOP10 = Analise_03['state'][0:10].values # elencando

# Tamanho
plt.figure( figsize=(12, 5) )

# Grafico
plt.title( 'Top 10 Estados com incêncios', loc='left', fontsize=14 )

# Loop
for Coluna in Lista_TOP10:
  
  # Filtrar o estado
  Filtro = dados.loc[ dados['state'] == Coluna ]

  # Agrupar os valores para sumarizar
  Analise_Local = Filtro.groupby( by=['year'] ).sum().reset_index() #agrupando os valores por ano

  # Plot
  sns.lineplot(data=Analise_Local, x='year', y='number', lw=2, alpha=0.85)

# Labels
plt.xlabel('Período')
plt.ylabel('Número de incêndios')

# Legenda
plt.legend( Lista_TOP10, bbox_to_anchor=(1, 0.7) );
```

```python
# listando os top 5 estados
top5 = Analise_03['state'][0:5].values

plt.figure( figsize=(12, 5) )

# Grafico
plt.title( 'Top 5 Estados com incêncios', loc='left', fontsize=14 )

for coluna in top5:
  filtro = dados.loc[dados['state'] == coluna] #omde coluna for igual ao estado, adiciona à variavel filtro
  ana_loc = filtro.groupby(by =['year']).sum().reset_index()
  sns.lineplot(data=ana_loc, x='year', y='number', lw=2, alpha=0.85)

plt.xlabel('Período')
plt.ylabel('Número de incêndios')

plt.legend( Lista_TOP10, bbox_to_anchor=(1, 0.7) );
```
